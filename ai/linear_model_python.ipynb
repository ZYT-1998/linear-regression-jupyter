# Linear Regression Analysis in Python

This notebook fits a linear regression model using YearsExperience as the predictor variable and Salary as the response variable. It calculates the slope, intercept, Pearson correlation coefficient, and mean squared error.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error

dataset = pd.read_csv("regression_data-1.csv")

x = dataset["YearsExperience"].to_numpy()
y = dataset["Salary"].to_numpy()

slope, intercept, r_value, p_value, std_error = linregress(x, y)

y_pred = slope * x + intercept
mse = mean_squared_error(y, y_pred)

print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"Pearson correlation coefficient (r): {r_value:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
order = np.argsort(x)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, label="Observed data")
plt.plot(x[order], y_pred[order], label="Fitted line")

annotation = (
    f"y = {slope:.2f}x + {intercept:.2f}\n"
    f"r = {r_value:.2f}\n"
    f"MSE = {mse:.2f}"
)

plt.text(
    0.05,
    0.95,
    annotation,
    transform=plt.gca().transAxes,
    verticalalignment="top",
    bbox={"boxstyle": "round", "alpha": 0.8}
)

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs. Years of Experience")
plt.legend()
plt.show()
```

## Interpretation

The fitted slope is about 8285.29. Based on this model, one additional year of experience is associated with an estimated salary increase of about $8285.

The intercept is about 29203.52, representing the model's predicted salary at zero years of experience. This value should be interpreted carefully because the intercept is mainly used to position the regression line and may not represent an actual starting salary.

The Pearson correlation coefficient is about 0.886. This shows a strong positive linear association between experience and salary, although correlation alone does not prove that experience directly causes the salary increase.

The MSE is about 17.52 million squared salary units. It summarizes the squared prediction errors, so larger differences between actual and predicted salaries have a stronger effect on this value.
